# Factorial
#Factorial means multiplication of a number in decreasing order upto one. ... In mathematics, the factorial of a non-negative integer n, denoted by n!, is the product of all positive integers less than or equal to n. For example, 5 ! = 5 × 4 × 3 × 2 × 1 = 120

```python
def sum():
    number = int(input("Enter the number?"))
    factorial = 1
    if number<0:#if number is less than zero understand its mean minuts
        return 0
    elif number==0:#if number is equal to zero its means zero
        print("zero")
    else:
        for x in range(1,number+1):
            factorial = factorial * x
        print("The factorial",number,"is: ",factorial)
sum()
```
